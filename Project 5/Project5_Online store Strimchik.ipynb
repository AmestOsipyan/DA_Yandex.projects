### Проверим пропуски и обработаем их, в случае необходимости

```python
data.isna().sum()
```

```python
data_isna=data[data['name'].isna()]
display(data_isna)
```

Пропуски в столбцах name и genre обнаружены в незначительном количестве (0.012% от датафрейма) и в одних и тех же строках. Заполнить данные пропуски невозможно. Поэтому их целесобразно удалить.

```python
data = data.dropna(subset=['name'])
```

```python
data_isna2=data[data['year_of_release'].isna()]
display(data_isna2)
```

Количество пропусков в столбце `year_of_release` - 269, что составляет 1.6% от общего датафрейма. Данные в столбцах можно заполнить с помощью поиска по названию игр и данных о годах выходах игр по официальным открытым источникам. Но на это потребуется дополнительное время, что заметно увеличит общее время проведения данного анализа.
Так как число пропусков в процентом соотношении от общего объема данных составляет 1.6%, мы можем себе позволить удалить строки с данными пропусками. Это не окажет влияния на анализ данных.

```python
data=data.dropna(subset=['year_of_release'])
```

```python
data.isna().sum()
```

```python
user_tbd = data.query('user_score == "tbd"')
display(user_tbd)
```

Пропуски в остальных 2-х стобцах `critic_score`, `user_score` оставляем в исходном виде, не заменяем.
В столбце `user_score` встречается значение `tbd` (2376 строк): данное значение можем отнести к пропущенным значениям. Эта аббревиатура может означать *'to be decided'(будет определено), 'to be deleted' (будет удалено).*
Заменим данные значение `tbd` на значение `NaN`

```python
data['user_score'] = data['user_score'].replace('tbd', 'NaN')
```

В столбце `rating` указаны рейтинги от организации ESRB (англ. Entertainment Software Rating Board). Эта ассоциация определяет рейтинг компьютерных игр (присваивает им подходящую возрастную категорию) в США и Канаде. 

Чтобы мы могли в дальнейшем анализе учесть игры в том числе по регионам, которые не охватывает ESRB, **заменим пропуски в столбце `rating` аббревиатурой OR (other rating)**.

```python
data['rating'] = data['rating'].fillna('OR')
data['rating'].unique()
```

Причиной возникновения пропусков в этом столбце может послужить тот факт, что игра не была выпущена для рынка США и Канады, а была выпущна для других рынков, например, Японии.

 - `user_score`, из текстового типа object преобразовали в числовой с плавающей точкой float. Данные в столбце хранятся в формате десятичной дроби.

## Проверим гипотезы

### Гипотеза 1
**Нулевая гипотеза:** средние пользовательские рейтинги платформ Xbox One и PC равны. **Альтернативная гипотеза** - средние пользовательские рейтинги платформ XOne и PC не равны.

```python
xone = data_actual[data_actual['platform'] == 'XOne']['user_score'].dropna()
pc = data_actual[data_actual['platform'] == 'PC']['user_score'].dropna()

alpha = 0.1

results = st.ttest_ind(xone, pc)

print('p-значение:', results.pvalue)
if results.pvalue < alpha:
    print('Нулевая гипотеза отвергнута. Cредние пользовательские рейтинги платформ XOne и PC не равны')
else:
    print('Нулевая гипотеза не отвергнута')
```

**Нулевая гипотеза:** Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») равны. **Альтернативная гипотеза:** Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные.

```python
action = data_actual[data_actual['genre'] == 'Action']['user_score'].dropna()
sports = data_actual[data_actual['genre'] == 'Sports']['user_score'].dropna()

alpha = 0.1

results = st.ttest_ind(action, sports)

print('p-значение:', results.pvalue)
if results.pvalue < alpha:
    print('Нулевая гипотеза отвергнута. Cредние пользовательские рейтинги платформ Action и Sports разные')
else:
    print('Нулевая гипотеза не отвергнута')
```
